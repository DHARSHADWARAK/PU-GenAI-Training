```python
documents = [ 
   "The cat sits on the mat", 
   "A feline is resting on the rug", 
   "Dogs and cats are common pets", 
   "The dog plays in the garden", 
   "Pets like cats and dogs are friendly", 
   "A kitten is a young cat", 
   "Puppies are young dogs", 
   "Animals like feline and canine are mammals" 
] 

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(stop_words="english")
x=vectorizer.fit_transform(documents)

words=vectorizer.get_feature_names_out()
x
```

```python
from sklearn.decomposition import TruncatedSVD

svd=TruncatedSVD(n_components=2,random_state=42)
x_Red=svd.fit_transform(x)
x_Red
```

```python
from sklearn.preprocessing import normalize
word_vectors=normalize(svd.components_.T)
word_vectors
```

```python
query="cat"
query_idx=list(words).index(query)
query_vec=word_vectors[query_idx]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities=cosine_similarity([query_vec],word_vectors)[0]
similarities
```

```python
import numpy as np

top_indices=np.argsort(similarities)[::-1]

for i in top_indices[0:4]:
    print(words[i],similarities[i])
```
