# AI Shopping Assistant using Router Chain + Prompt Templates + Memory

This notebook demonstrates the implementation from `shopping_assistant.py`.

It includes:
- Router logic
- Product recommendation chain
- Customer support chain
- Return and refund FAQ chain
- Large FAQ chain with example selection
- Structured output chain
- Bonus upselling chain
- Buffer, window, summary, and entity memory concepts

## Architecture

```text
Customer -> ShoppingRouterChain -> Specialized Chain -> Chain Memory -> Assistant Response

Routes:
- Product Recommendation: PromptTemplate + Entity Memory
- Customer Support: ChatPromptTemplate + Buffer Memory
- Return and Refund FAQ: FewShotPromptTemplate + Window Memory
- Large FAQ: Example Selector + Summary Memory
- Structured Output: JSON Prompt + Optional Memory
- Upselling: PromptTemplate + Entity + Window Memory
```

````python
from shopping_assistant import AIShoppingAssistant

assistant = AIShoppingAssistant(use_openai=False)
````

````python
queries = [
    "Suggest headphones under 2000",
    "Where is my order ORD12345?",
    "Can I return electronics?",
    "Is COD available and what about warranty?",
    "Give product details in JSON format for a phone under 20000",
    "I bought a laptop, suggest an accessory",
]

for query in queries:
    result = assistant.ask(query)
    print("\nUser:", query)
    print("Route:", result["route"])
    print("Prompt:", result["prompt_type"])
    print("Memory:", result["memory_type"])
    print("Assistant:", result["answer"])
````

## Prompt Comparison

| Prompt Type | Best Use |
|---|---|
| PromptTemplate | Simple recommendation |
| ChatPromptTemplate | Conversation |
| FewShotPromptTemplate | FAQ learning |
| Example Selector | Large FAQ system |
| Structured Output Prompt | JSON/API output |

## Memory Comparison

| Memory Type | Best Use |
|---|---|
| Buffer | Full conversation |
| Window | Recent chat |
| Summary | Long history |
| Entity | Important details |
