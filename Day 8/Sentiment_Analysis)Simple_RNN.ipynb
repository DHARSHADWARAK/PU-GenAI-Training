```python
docs = [
    "the sun is shining",
    "rain is falling heavily",
    "birds are singing in the morning",
    "the cat is sleeping on the mat",
    "dogs are playing in the park",
    "flowers are blooming everywhere",
    "the wind is blowing strongly",
    "children are laughing loudly",
    "the moon is bright tonight",
    "stars are twinkling in the sky"
]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences = pad_sequences(sequences,padding='post')
sequences
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding

model = Sequential()
model.add(Embedding(17,output_dim=2,input_length=5))

model.summary()
```

```python
model.compile('adam','accuracy')
```

```python
pred = model.predict(sequences)
print(pred)
```

```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=2, input_length=50))

model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train, y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.datasets import imdb

word_index = imdb.get_word_index()

def text_to_sequence(text, word_index, maxlen=50):
    seq = [word_index.get(w.lower(), 2) + 3 for w in text.split()]
    return pad_sequences([seq], maxlen=maxlen, padding='post')

def predict_sentiment(text):
    seq = text_to_sequence(text, word_index)
    pred = model.predict(seq)
    return "Positive" if pred[0][0] > 0.5 else "Negative"

# Example usage
print(predict_sentiment("I loved the movie, it was fantastic!"))
print(predict_sentiment("The movie was boring and too long."))
```
