```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Embedding


text = "hello world"
chars = sorted(list(set(text)))
char_to_int = {c: i for i, c in enumerate(chars)}
int_to_char = {i: c for i, c in enumerate(chars)}


seq_length = 3
dataX, dataY = [], []
for i in range(len(text) - seq_length):
    seq_in = text[i:i+seq_length]
    seq_out = text[i+seq_length]
    dataX.append([char_to_int[ch] for ch in seq_in])
    dataY.append(char_to_int[seq_out])

X = np.array(dataX)
y = tf.keras.utils.to_categorical(dataY, num_classes=len(chars))


model = Sequential()
model.add(Embedding(input_dim=len(chars), output_dim=10, input_length=seq_length))
model.add(SimpleRNN(20, return_sequences=False))
model.add(Dense(len(chars), activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


model.fit(X, y, epochs=200, verbose=0)



```

```python
seed = input("Enter The Seed Value")
print("\nSeed:", seed)
for _ in range(10):
    x_pred = np.array([[char_to_int[ch] for ch in seed[-seq_length:]]])
    pred = model.predict(x_pred, verbose=0)
    next_index = np.argmax(pred)
    next_char = int_to_char[next_index]
    seed += next_char

print("Generated text:", seed)
```

```python
import numpy as np


timesteps = 4   # sequence length
input_dim = 5   # input features
hidden_dim = 3  # hidden units
output_dim = 2  # output classes

x = np.random.randn(timesteps, input_dim)

Wx = np.random.randn(input_dim, hidden_dim)   # input вҶ’ hidden
Wh = np.random.randn(hidden_dim, hidden_dim)  # hidden вҶ’ hidden
bh = np.zeros((hidden_dim,))                  # hidden bias

Wy = np.random.randn(hidden_dim, output_dim)  # hidden вҶ’ output
by = np.zeros((output_dim,))                  # output bias


h = np.zeros((hidden_dim,))
hs = []
for t in range(timesteps):
    h = np.tanh(np.dot(x[t], Wx) + np.dot(h, Wh) + bh)
    hs.append(h)


y = np.dot(h, Wy) + by

print("Hidden states (one per timestep):")
for i, state in enumerate(hs):
    print(f"t={i+1}: {state}")

print("\nFinal output:", y)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense


def prepare_data(text, seq_length=40):
    chars = sorted(list(set(text)))
    char_to_int = {c: i for i, c in enumerate(chars)}
    int_to_char = {i: c for i, c in enumerate(chars)}
    vocab_size = len(chars)

    dataX, dataY = [], []
    for i in range(len(text) - seq_length):
        seq_in = text[i:i+seq_length]
        seq_out = text[i+seq_length]
        dataX.append([char_to_int[ch] for ch in seq_in])
        dataY.append(char_to_int[seq_out])

    X = np.zeros((len(dataX), seq_length, vocab_size), dtype=np.float32)
    for i, seq in enumerate(dataX):
        for t, idx in enumerate(seq):
            X[i, t, idx] = 1.0

    y = tf.keras.utils.to_categorical(dataY, num_classes=vocab_size)
    return X, y, vocab_size, char_to_int, int_to_char


def build_model(seq_length, vocab_size, hidden_units=128):
    model = Sequential()
    model.add(SimpleRNN(hidden_units, input_shape=(seq_length, vocab_size)))
    model.add(Dense(vocab_size, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model


def generate_text(model, seed, char_to_int, int_to_char, vocab_size, gen_length=200):
    pattern = [char_to_int[c] for c in seed]
    text = seed
    for _ in range(gen_length):
        x = np.zeros((1, len(pattern), vocab_size))
        for t, idx in enumerate(pattern):
            x[0, t, idx] = 1.0
        prediction = model.predict(x, verbose=0)
        index = np.argmax(prediction)
        result = int_to_char[index]
        text += result
        pattern = pattern[1:] + [index]
    return text


text = """To be, or not to be, that is the question:
Whether 'tis nobler in the mind to suffer
The slings and arrows of outrageous fortune..."""

seq_length = 40
X, y, vocab_size, char_to_int, int_to_char = prepare_data(text, seq_length)

model = build_model(seq_length, vocab_size, hidden_units=128)
model.fit(X, y, epochs=50, verbose=1)

seed_text = "To be, or not to be, that is the "
generated = generate_text(model, seed_text, char_to_int, int_to_char, vocab_size, gen_length=200)

print("\nGenerated Text:\n", generated)
```

```python
import tensorflow as tf
import numpy as np

path_to_file = tf.keras.utils.get_file("shakespeare.txt", "https://storage.googleapis.com/download.tensorflow.org/data/shakespeare.txt")
text = open(path_to_file, 'rb').read().decode(encoding='utf-8')
vocab = sorted(set(text))
char2idx = {u: i for i, u in enumerate(vocab)}
idx2char = np.array(vocab)
text_as_int = np.array([char2idx[c] for c in text])

seq_length = 100
char_dataset = tf.data.Dataset.from_tensor_slices(text_as_int)
sequences = char_dataset.batch(seq_length+1, drop_remainder=True)

def split_input_target(chunk):
    return chunk[:-1], chunk[1:]

dataset = sequences.map(split_input_target)
BATCH_SIZE = 64
BUFFER_SIZE = 10000
dataset = dataset.shuffle(BUFFER_SIZE).batch(BATCH_SIZE, drop_remainder=True)

vocab_size = len(vocab)
embedding_dim = 256
rnn_units = 512

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=seq_length),
    tf.keras.layers.LSTM(rnn_units, return_sequences=True, stateful=False, recurrent_initializer='glorot_uniform'),
    tf.keras.layers.Dense(vocab_size)
])

model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))
model.fit(dataset, epochs=5)
```
