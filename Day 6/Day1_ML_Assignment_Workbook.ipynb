### Task 14 (8 pts): Overfitting vs Underfitting Investigation
The following simulated scenario shows different models. Your task:

1. Run the cell to generate results
2. Identify which model is overfitting, which is underfitting, which is just right
3. Explain WHY for each


```python
# Pre-built scenario — read results and diagnose
from sklearn.preprocessing import PolynomialFeatures

np.random.seed(10)
X_ov = np.sort(np.random.rand(80, 1) * 10, axis=0)
y_ov = 1.5*X_ov.ravel() + np.random.randn(80) * 2.5

X_tr_ov, X_te_ov, y_tr_ov, y_te_ov = train_test_split(X_ov, y_ov, test_size=0.3, random_state=42)

results_ov = {}
for deg, name in [(1, 'Model A (degree=1)'), (2, 'Model B (degree=2)'), (12, 'Model C (degree=12)')]:
    from sklearn.pipeline import Pipeline
    p = Pipeline([('poly', PolynomialFeatures(deg)), ('lr', LinearRegression())])
    p.fit(X_tr_ov, y_tr_ov)
    tr2 = r2_score(y_tr_ov, p.predict(X_tr_ov))
    te2 = r2_score(y_te_ov, p.predict(X_te_ov))
    results_ov[name] = {'Train R²': round(tr2,4), 'Test R²': round(te2,4), 'Gap': round(tr2-te2,4)}
    print(f'{name:<28} Train R²={tr2:.4f}   Test R²={te2:.4f}   Gap={tr2-te2:.4f}')
```
