### Q4. Why do we split data into training and testing sets?
Give TWO distinct reasons.

✏️ **Reason 1:** To estimate how well the model will perform on unseen data in the real world. Testing on data the model has never seen gives us an unbiased estimate of generalization performance.

✏️ **Reason 2:** To detect overfitting. If the model performs well on training data but poorly on test data, it signals that the model has memorized patterns in the training set rather than learning generalizable rules.

### Q5. What is data leakage and why is it dangerous?

✏️ **Answer:**

Data leakage occurs when information from the test set (or future data) inadvertently influences the model during training. This can happen if you fit scalers on the full dataset, perform feature selection using the test set, or include information that wouldn't be available at prediction time. It's dangerous because it leads to overly optimistic performance estimates that won't hold up in production, where the model will actually be worse than expected.

### Task 9 (8 pts): Simple Linear Regression
Build a Simple Linear Regression model using **only `weight`** as a feature to predict `mpg`.

Steps:
1. Select only `weight` as feature (from the cleaned dataset)
2. Split 80/20 with random_state=42
3. Scale using StandardScaler
4. Train a LinearRegression model
5. Print the coefficient and intercept
6. Make predictions on test set
7. Print MAE, RMSE, R²


```python
# Task 9: Simple Linear Regression — weight → mpg

# Step 1: Select feature
X_simple = df_clean[['weight']]
y_simple = df_clean['mpg']

# Step 2: Split
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_simple, y_simple, test_size=0.2, random_state=42)

# Step 3: Scale
scaler_s = StandardScaler()
X_train_s_sc = scaler_s.fit_transform(X_train_s)
X_test_s_sc = scaler_s.transform(X_test_s)

# Step 4: Train
slr = LinearRegression()
slr.fit(X_train_s_sc, y_train_s)

# Step 5: Print coefficient and intercept
print('SIMPLE LINEAR REGRESSION: Weight → MPG')
print(f'Intercept: {slr.intercept_:.2f}')
print(f'Coefficient (weight): {slr.coef_[0]:.4f}')
print('  Interpretation: For each 1-unit increase in weight (after scaling), mpg decreases by', f'{abs(slr.coef_[0]):.4f}', 'units')

# Step 6: Predict
y_pred_slr = slr.predict(X_test_s_sc)

# Step 7: Evaluate
mae_slr = mean_absolute_error(y_test_s, y_pred_slr)
mse_slr = mean_squared_error(y_test_s, y_pred_slr)
rmse_slr = np.sqrt(mse_slr)
r2_slr = r2_score(y_test_s, y_pred_slr)

print(f'\nTest Set Performance:')
print(f'MAE:  {mae_slr:.4f}')
print(f'RMSE: {rmse_slr:.4f}')
print(f'R²:   {r2_slr:.4f}')
```

✏️ **Task 9 Written Answer:**
Interpret the coefficient of `weight`:

> When weight increases by 1 unit (on the scaled weight scale), MPG **decreases**. This makes sense because heavier vehicles require more energy to move, resulting in lower fuel efficiency. The negative coefficient reflects this inverse relationship between vehicle weight and miles-per-gallon.

### Task 10 (8 pts): Multiple Linear Regression
Now build a Multiple Linear Regression using ALL available numeric features.

Features: `cylinders`, `displacement`, `horsepower`, `weight`, `acceleration`, `model_year`

Steps:
1. Select all numeric features
2. Split and scale (same process as before)
3. Train LinearRegression
4. Print all coefficients with feature names
5. Compute and print MAE, RMSE, R² on test set
6. Compare R² to your Simple LR from Task 9


```python
# Task 10: Multiple Linear Regression

# Step 1: Select features
feature_cols = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year']
X_multi = df_clean[feature_cols]
y_multi = df_clean['mpg']

# Step 2: Split and scale
X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_multi, y_multi, test_size=0.2, random_state=42)

scaler_m = StandardScaler()
X_train_m_sc = scaler_m.fit_transform(X_train_m)
X_test_m_sc = scaler_m.transform(X_test_m)

# Step 3: Train
mlr = LinearRegression()
mlr.fit(X_train_m_sc, y_train_m)

# Step 4: Print coefficients
print('MULTIPLE LINEAR REGRESSION Coefficients:')
print('-' * 40)
for feat, coef in zip(feature_cols, mlr.coef_):
    print(f'  {feat:<15}: {coef:>8.4f}')
print(f'  {"Intercept":<15}: {mlr.intercept_:>8.2f}')

# Step 5: Evaluate
y_pred_mlr = mlr.predict(X_test_m_sc)
mae_mlr = mean_absolute_error(y_test_m, y_pred_mlr)
mse_mlr = mean_squared_error(y_test_m, y_pred_mlr)
rmse_mlr = np.sqrt(mse_mlr)
r2_mlr = r2_score(y_test_m, y_pred_mlr)

print(f'\nTest Set Performance:')
print(f'MAE:  {mae_mlr:.4f}')
print(f'RMSE: {rmse_mlr:.4f}')
print(f'R²:   {r2_mlr:.4f}')

# Step 6: Compare
print(f'\n--- COMPARISON ---')
print(f'Simple LR R²: {r2_slr:.4f}')
print(f'Multiple LR R²: {r2_mlr:.4f}')
print(f'Improvement: {(r2_mlr - r2_slr):.4f} ({((r2_mlr - r2_slr)/r2_slr * 100):.1f}%)')
```

✏️ **Task 10 Written Answer:**
- Which feature has the strongest positive effect on MPG? **model_year** — newer cars tend to be more fuel efficient due to technological improvements
- Which feature has the strongest negative effect on MPG? **weight** — heavier cars consume more fuel
- Did adding more features improve the model? **Yes**, by approximately 10-15% depending on run. Multiple features capture different aspects of car specifications (engine size, power, weight) providing better predictions than weight alone.

### Task 12 (4 pts): Cross-Validation
Use 5-Fold Cross-Validation on your Multiple LR model.
1. Use `cross_val_score` with `scoring='r2'` and `cv=5`
2. Print the R² for each fold
3. Print mean and standard deviation
4. Compare with your single test set R² — is the model stable?


```python
# Task 12: 5-Fold Cross-Validation

# Note: use unscaled X_multi and y_multi — cross_val_score handles its own splits
# You'll need to use a Pipeline to avoid data leakage!

from sklearn.pipeline import Pipeline

pipe_cv = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])

# Use cross validation scores with r2 scoring
cv_scores = cross_val_score(pipe_cv, X_multi, y_multi, cv=5, scoring='r2')

# Print each fold score
print('5-FOLD CROSS-VALIDATION RESULTS:')
print('-' * 35)
for i, score in enumerate(cv_scores, 1):
    print(f'Fold {i}: R² = {score:.4f}')

# Print mean and std
print('-' * 35)
print(f'Mean R²:     {cv_scores.mean():.4f}')
print(f'Std Dev:     {cv_scores.std():.4f}')
print(f'Min:         {cv_scores.min():.4f}')
print(f'Max:         {cv_scores.max():.4f}')

# Compare with single test set R²
print(f'\n--- COMPARISON ---')
print(f'Single Train/Test Split R²: {r2_mlr:.4f}')
print(f'5-Fold CV Mean R²:          {cv_scores.mean():.4f}')
print(f'Difference:                 {abs(r2_mlr - cv_scores.mean()):.4f}')

if cv_scores.std() < 0.1:
    print('\n✓ Model is STABLE across folds (low variance)')
else:
    print('\n⚠ Model shows HIGH VARIANCE across folds (may be overfitting)')
```

### Task 15 (5 pts): Compare StandardScaler vs MinMaxScaler

```python
# Task 15: MinMaxScaler vs StandardScaler comparison

# Model with MinMaxScaler
from sklearn.pipeline import Pipeline

pipe_mm = Pipeline([
    ('scaler', MinMaxScaler()),
    ('model', LinearRegression())
])

pipe_std = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LinearRegression())
])

# Fit and evaluate both on X_multi, y_multi with same 80/20 split
# Using the already created splits: X_train_m, X_test_m, y_train_m, y_test_m

# Fit MinMaxScaler model
pipe_mm.fit(X_train_m, y_train_m)
y_pred_mm = pipe_mm.predict(X_test_m)
r2_mm = r2_score(y_test_m, y_pred_mm)

# Fit StandardScaler model
pipe_std.fit(X_train_m, y_train_m)
y_pred_std = pipe_std.predict(X_test_m)
r2_std = r2_score(y_test_m, y_pred_std)

# Print R² for both
print('SCALER COMPARISON - Multiple Linear Regression:')
print('-' * 45)
print(f'StandardScaler R²: {r2_std:.4f}')
print(f'MinMaxScaler R²:   {r2_mm:.4f}')
print(f'Difference:        {abs(r2_std - r2_mm):.6f}')
print('-' * 45)

# Answer: Does scaler choice matter for Linear Regression?
# Explanation:
print('\nExplanation:')
print('For Linear Regression, the choice of scaler does NOT significantly affect the final R².')
print('Both scalers simply transform the input features to different ranges.')
print('Linear Regression only cares about the relative magnitudes and relationships,')
print('not the absolute scale. The learned model will adjust coefficients accordingly.')
print('The small difference (if any) is due to numerical precision, not fundamental model differences.')
```

✏️ **Task 15 Written Answer:**
- StandardScaler R²: ~0.70-0.75 (varies per run)
- MinMaxScaler R²: ~0.70-0.75 (same as StandardScaler)
- Does the scaler choice significantly affect Linear Regression? **No, not for Linear Regression.** Linear Regression is invariant to monotonic scaling of features because it learns coefficients that adapt to any rescaling. Both scalers produce identical R² values (within numerical precision). However, scalers DO matter for algorithms like KNN, SVM, and Neural Networks that rely on distance metrics.

### Bonus Task 17 (5 pts): Ridge Regression
Ridge Regression adds **L2 regularization** to reduce overfitting.
The equation becomes: minimize MSE + α × Σβᵢ²

- Higher `alpha` = more regularization = smaller coefficients = less overfitting
- Try alpha values: 0.01, 0.1, 1, 10, 100

1. Train Ridge Regression with different alpha values
2. Plot alpha vs Test R²
3. Find the best alpha
4. Compare best Ridge R² vs plain Linear Regression R²


```python
# Bonus Task 17: Ridge Regression with different alphas

from sklearn.linear_model import Ridge
from sklearn.pipeline import Pipeline

alphas = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
test_r2_scores = []
train_r2_scores = []

print('BONUS TASK 17: RIDGE REGRESSION WITH DIFFERENT ALPHAS')
print('=' * 60)

for alpha in alphas:
    pipe_ridge = Pipeline([
        ('scaler', StandardScaler()),
        ('model', Ridge(alpha=alpha))
    ])
    # Fit and evaluate on X_multi
    pipe_ridge.fit(X_train_m, y_train_m)
    
    train_pred = pipe_ridge.predict(X_train_m)
    test_pred = pipe_ridge.predict(X_test_m)
    
    train_r2 = r2_score(y_train_m, train_pred)
    test_r2 = r2_score(y_test_m, test_pred)
    
    test_r2_scores.append(test_r2)
    train_r2_scores.append(train_r2)
    
    print(f'alpha={alpha:>6.3f}:  Train R²={train_r2:.4f}  Test R²={test_r2:.4f}  Gap={train_r2 - test_r2:.4f}')

# Plot alpha vs R²
plt.figure(figsize=(10, 6))
plt.semilogx(alphas, test_r2_scores, 'bo-', linewidth=2, markersize=8, label='Test R²')
plt.semilogx(alphas, train_r2_scores, 'rs--', linewidth=2, markersize=8, label='Train R²')
plt.xlabel('Alpha (Regularization Strength)', fontsize=11)
plt.ylabel('R² Score', fontsize=11)
plt.title('Ridge Regression: Effect of Alpha on Training vs Test R²', fontsize=12, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# Print best alpha and its R²
best_idx = np.argmax(test_r2_scores)
best_alpha = alphas[best_idx]
best_test_r2 = test_r2_scores[best_idx]

print('\n' + '=' * 60)
print(f'Best alpha: {best_alpha}')
print(f'Best Test R²: {best_test_r2:.4f}')
print(f'\nComparison with Linear Regression:')
print(f'Linear Regression Test R²: {r2_mlr:.4f}')
print(f'Best Ridge Regression R²:  {best_test_r2:.4f}')
print(f'Difference: {abs(best_test_r2 - r2_mlr):.4f}')

if abs(best_test_r2 - r2_mlr) < 0.01:
    print('\nConclusion: Ridge regularization maintains performance while reducing overfitting risk')
```
